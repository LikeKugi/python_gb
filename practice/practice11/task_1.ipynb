```python
ashechka = [1743, 1648, 1650, 1622, 1581, 1490]
bshechka = [743, 648, 711, 780, 805, 846]
```

```python
ab_speed = list(zip(ashechka, bshechka))
```

```python
print(ab_speed)
```

```python
ab_speed = list(map(lambda x: x[0]-x[1], ab_speed))
```

```python
print(ab_speed)
```

```python
diffs = []
for i, el in enumerate(ab_speed[:-1]):
    diffs.append(ab_speed[i]- ab_speed[i+1])
    
print( diffs)
```

```python
a_speed = []
for i, el in enumerate(ashechka[:-1]):
    a_speed.append(ashechka[i] - ashechka[i+1])
print(a_speed)
```

```python
a_mean = int(sum(a_speed) // len(a_speed))
print(a_mean)
```

```python
b_speed = []
for i, el in enumerate(bshechka[:-1]):
    b_speed.append(bshechka[i]-bshechka[i+1])
print(b_speed)
```

```python
b_mean = int(sum(b_speed) // len(b_speed))
print(b_mean)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(a_speed)
```

```python
i = len(ashechka) - 1
while ashechka[i] > bshechka[i]:
    
    ashechka.append(ashechka[i-1]-a_mean)
    bshechka.append(bshechka[i-1]-b_mean)
    
    i+=1
    
print(ashechka)
print(bshechka)
```

```python
plt.plot(ashechka, color='r', label='a_team')
plt.plot(bshechka, color='g', label='b_team')
```
