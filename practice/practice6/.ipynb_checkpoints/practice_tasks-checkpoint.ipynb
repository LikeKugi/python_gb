Задача 3. Имеется 1000 рублей. Бык стоит – 100
рублей, корова – 50 рублей, телёнок – 5 рублей.
Сколько быков, коров и телят можно купить на все
эти деньги, если всего надо купить 100 голов
скота?

```python
total = 1000
bull_cost = 100
cow_cost = 50
child_cost = 5

for x in range(10):
    for y in range(20):
        for z in range(200):
            if (x*bull_cost + y*cow_cost + z*child_cost == total) and x+y+z==100:
                print(f'bulls: {x}')
                print(f'cows: {y}')
                print(f'children: {z}')
```

Задача 1. Дан список, состоящий из случайных целых чисел. Дано число M. Выведите случайное подмножество список, сумма элементов в котором не превосходит M.

```python
numbers = [RI(0,55) for _ in range(10)]
print(numbers)
M = 50
sublist = []
nl = numbers.copy()
while len(nl) > 0:
    a = nl.pop(RI(0,len(nl)-1))
    if a < M and sum(sublist) + a < M:
        sublist.append(a)
        
print(sublist)
    
```

```python
import numpy as np
n = int(input('rows: '))
m = int(input('columns: '))
max_value = int(input('max value of matrix: '))
matrix = np.random.randint(max_value, size=(n,m))
print(matrix)
```

```python
sides = set(el for el in matrix.reshape(-1) if el > 0)
print(sides)
```

```python
import math
sides = list(sides)
r = len(sides)
variants = []
for x in range(r-2):
    print(x)
    for y in range(x+1, r-1):
        print(y)
        for z in range(y+1, r):
            print(z)
            print(sides[x],sides[y],sides[z])
            if sum([sides[x],sides[y],sides[z]]) - max(sides[x],sides[y],sides[z]) > max(sides[x],sides[y],sides[z]):
                variants.append([sides[x],sides[y],sides[z]])

print(variants)
```

```python
def get_square(sds: list) -> int:
    p = sum(sds)/2
    S = math.sqrt(p*(p-sds[0])*(p-sds[1])*(p-sds[2]))
    return S
```

```python
print(max(variants,key=get_square))
```

Задача 3. Дан отсортированный по возрастанию двумерный массив Matrix.На вход подаётся число A. Напишите метод, реализующий поиск в массиве элементов, равных A. Выведите координаты элемента (при выводе нумерация строк и столбцов считается с единицы).

```python
import numpy as np
n = int(input('rows: '))
m = int(input('columns: '))
max_value = int(input('max value of matrix: '))
matrix = np.random.randint(max_value, size=(n,m))
print(matrix)
```

```python
matrix = matrix.reshape(-1)
print(matrix)
```

```python
matrix = matrix.reshape(m,n)
```

```python
A = int(input('number to search: '))
```

```python
def find_number(num:int, matrix, start=0, end=-1) -> tuple:
    if end == -1:
        end = matrix.shape[0]
    flag = True
    ctr = 0
    while flag:
        
        if ctr == 100:
            break
            
        mid = (end + start) // 2
        
        print(num, mid, matrix[mid,-1])
        
        if matrix[mid, 0] <= num <= matrix[mid, -1]:
            s = 0; e = matrix.shape[1]
            
            while e - s > 0:
                m = (e+s)//2
                
                if matrix[mid, m] == num:
                    return (mid, m)
                
                elif matrix[mid,m] < num:
                    s = m+1
                    
                else: 
                    e = m-1
                    
        elif matrix[mid, -1] < num:
            start = mid + 1
            print(f'start: {start}; end: {end}; mid: {mid}')
        else:
            end = mid - 1
        ctr += 1
        
    
```
